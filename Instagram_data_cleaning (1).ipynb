Let's analyze now the outliers on the numeric columns, that are: 'reach', 'impressions', 'saves', 'link_clicks', '3-Sec-View', '100%_visualized', 'comments', 'reactions' and 'shares'.

To do so, let's look for values that have 3 or more standart variations from the mean.

```python
num_columns=['reach', 'impressions', 'saves', 'link_clicks', '3-Sec-View', '100%_visualized', 'comments', 'reactions', 'shares']
for column in num_columns:
    mean=df[column].mean()
    std=df[column].std()
    if std==0:std=0.1
    for i in df[column]:
        if ((int(i)-mean)/std)>3:
            print(f'{i} in {column} from the post {df.post_name[df[column]==i].values[0]} is an outlier, standing {((i-mean)/std)} standart deviations from the mean that is {mean}. \n')
```
